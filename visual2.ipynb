```python
# read the multiple json files
# and visualize the data

import json
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
import plotly.graph_objects as go


def read_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
        # contains several keys
        # extract the key 'server'
        data = data['server']
        # data1 = data['best_assignment']
    return data

def read_json_best(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)
        # contains several keys
        # extract the key 'server'
        # data = data['server']
        data1 = data['best_assignment']
    return data1

def plot_acc(data, title):
    plt.figure()
    plt.title(title)
    plt.xlabel('rounds')
    plt.ylabel('accuracy')
    plt.plot(data)
    # plt.savefig(save_path)
    # plt.close()

def plot_f1(data, title):
    plt.figure()
    plt.title(title)
    plt.xlabel('rounds')
    plt.ylabel('F1 score')
    plt.plot(data)
    # plt.savefig(save_path)
    # plt.close()

```

```python

path = 'log/' #'log/log_fash_nogroup/'
# path='/'
file_list = os.listdir(path)
file_list = [file_path for file_path in file_list if file_path[-5:] == '.json' and 'jammer' in file_path and 'central' not in file_path]
fig_acc = go.Figure()
fig_f1 = go.Figure()
for file_path in file_list:
    data = read_json(path+file_path)
    data = np.array(data)
    fig_acc.add_trace(go.Scatter(y=data[:,0],name=file_path[:30]))
    fig_f1.add_trace(go.Scatter(y=data[:,1],name=file_path[:30]))
    
fig_acc.update_layout(title='Accuracy', xaxis_title='rounds', yaxis_title='accuracy')
fig_f1.update_layout(title='F1 score', xaxis_title='rounds', yaxis_title='F1 score')
# fig_acc.write_html(path + 'acc.html')
# fig_f1.write_html(path + 'f1.html')
fig_acc.show()
fig_f1.show()
```
